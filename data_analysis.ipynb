```python
# Step A â Data preparation and merge (2000â2023)
import pandas as pd
import numpy as np

# Start from already-created dataframes: art_coverage_SA, hiv_estimates_SA, aids_deaths_SA
# 1) Clean and coerce types
art_clean = art_coverage_SA.copy()
art_clean['Year'] = pd.to_numeric(art_clean['Year'], errors='coerce')
art_clean['Total ART coverage'] = pd.to_numeric(art_clean.get('Total ART coverage'), errors='coerce')

prev_clean = hiv_estimates_SA[['Year', 'HIV prevalence (%)']].copy()
prev_clean['Year'] = pd.to_numeric(prev_clean['Year'], errors='coerce')
prev_clean['HIV prevalence (%)'] = pd.to_numeric(prev_clean['HIV prevalence (%)'], errors='coerce')

# deaths may be like '15 000' â strip spaces/commas then to numeric
_deaths = aids_deaths_SA[['Year', 'deaths_estimate']].copy()
_deaths['Year'] = pd.to_numeric(_deaths['Year'], errors='coerce')
_deaths['deaths_estimate'] = (
	_deaths['deaths_estimate']
	.astype(str)
	.str.replace(r'[ ,]', '', regex=True)
)
_deaths['deaths_estimate'] = pd.to_numeric(_deaths['deaths_estimate'], errors='coerce')

# 2) Merge to a single table and restrict to 2000â2023 window
merged = (
	prev_clean
	.merge(art_clean[['Year', 'Total ART coverage']], on='Year', how='left')
	.merge(_deaths, on='Year', how='left')
)
merged = merged[(merged['Year'] >= 2000) & (merged['Year'] <= 2023)]
merged = merged.sort_values('Year').reset_index(drop=True)

# 3) Overlap rows where all three series are present (for correlation/regression)
analysis_df = merged.dropna(subset=['HIV prevalence (%)', 'Total ART coverage', 'deaths_estimate']).copy()

print('Merged (tail):')
display(merged.tail(12))
print('\nAnalysis subset (rows with all variables present):')
display(analysis_df)
```

```python
# Step C â Correlations and regression model comparisons
from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
import numpy as np

# Use rows with complete data
if analysis_df.empty:
	print('No overlapping years across all variables. Correlations/regressions skipped.')
else:
	# Correlations
	corr_art_prev = analysis_df[['Total ART coverage', 'HIV prevalence (%)']].corr(method='pearson').iloc[0,1]
	corr_art_deaths = analysis_df[['Total ART coverage', 'deaths_estimate']].corr(method='pearson').iloc[0,1]
	print(f"Pearson r (ART vs HIV prevalence): {corr_art_prev:.3f}")
	print(f"Pearson r (ART vs AIDS deaths): {corr_art_deaths:.3f}")

	# Prepare X and y for models
	X = analysis_df[['Total ART coverage']].values.astype(float)
	y_prev = analysis_df['HIV prevalence (%)'].values.astype(float)
	y_deaths = analysis_df['deaths_estimate'].values.astype(float)

	def fit_and_compare_models(X, y, label):
		# Linear
		lin = LinearRegression().fit(X, y)
		yhat_lin = lin.predict(X)
		r2_lin = r2_score(y, yhat_lin)

		# Logarithmic (y ~ a + b*ln(X)), need X>0
		mask_pos = (X.flatten() > 0)
		if mask_pos.sum() >= 2:
			X_log = np.log(X[mask_pos])
			y_log = y[mask_pos]
			lin_log = LinearRegression().fit(X_log, y_log)
			yhat_log = lin_log.predict(X_log)
			r2_log = r2_score(y_log, yhat_log)
		else:
			r2_log = np.nan

		# Exponential (ln(y) ~ a + b*X), need y>0
		mask_ypos = (y > 0)
		if mask_ypos.sum() >= 2:
			lin_exp = LinearRegression().fit(X[mask_ypos], np.log(y[mask_ypos]))
			yhat_exp = np.exp(lin_exp.predict(X))
			r2_exp = r2_score(y[mask_ypos], yhat_exp[mask_ypos])
		else:
			r2_exp = np.nan

		# Quadratic
		pf = PolynomialFeatures(degree=2, include_bias=False)
		X_quad = pf.fit_transform(X)
		lin_quad = LinearRegression().fit(X_quad, y)
		yhat_quad = lin_quad.predict(X_quad)
		r2_quad = r2_score(y, yhat_quad)

		print(f"\nR^2 for {label}:")
		print(f"  Linear:       {r2_lin:.3f}")
		print(f"  Logarithmic:  {r2_log if not np.isnan(r2_log) else 'NA'}")
		print(f"  Exponential:  {r2_exp if not np.isnan(r2_exp) else 'NA'}")
		print(f"  Quadratic:    {r2_quad:.3f}")

	fit_and_compare_models(X, y_prev, 'HIV prevalence vs ART')
	fit_and_compare_models(X, y_deaths, 'AIDS deaths vs ART')
```

```python
# Step D â Advanced analyses: lag, piecewise, and rates of change
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

if analysis_df.empty:
	print('No overlapping years across all variables. Advanced analyses skipped.')
else:
	# 1) Lag analysis: shift ART forward by k years (does ART at t relate to outcomes at t+k?)
	def lag_correlation(df, lag_years):
		z = df[['Year', 'Total ART coverage', 'HIV prevalence (%)', 'deaths_estimate']].copy()
		z['Total_lag'] = z['Total ART coverage'].shift(periods=-lag_years)  # ART at t matched to outcomes at t+lag
		z = z.dropna(subset=['Total_lag'])
		corr_prev = z[['Total_lag', 'HIV prevalence (%)']].corr().iloc[0,1]
		corr_deaths = z[['Total_lag', 'deaths_estimate']].corr().iloc[0,1]
		return corr_prev, corr_deaths

	for k in [0, 1, 2]:
		cp, cd = lag_correlation(analysis_df, k)
		print(f"Lag {k}y â r(ARTâPrev)={cp:.3f}, r(ARTâDeaths)={cd:.3f}")

	# 2) Piecewise analysis: pre-2004 vs post-2004
	pre = analysis_df[analysis_df['Year'] < 2004]
	post = analysis_df[analysis_df['Year'] >= 2004]
	def slope_vs_art(df, y_col):
		if len(df) < 2:
			return np.nan
		m = LinearRegression().fit(df[['Total ART coverage']], df[y_col])
		return float(m.coef_[0])
	print('\nPiecewise slopes (delta outcome per 1% ART):')
	print('  HIV prevalence pre-2004:', slope_vs_art(pre, 'HIV prevalence (%)'))
	print('  HIV prevalence post-2004:', slope_vs_art(post, 'HIV prevalence (%)'))
	print('  AIDS deaths pre-2004:', slope_vs_art(pre, 'deaths_estimate'))
	print('  AIDS deaths post-2004:', slope_vs_art(post, 'deaths_estimate'))

	# 3) Rates of change per year
	rate_df = merged.sort_values('Year')[['Year', 'HIV prevalence (%)', 'Total ART coverage', 'deaths_estimate']].copy()
	for col in ['HIV prevalence (%)', 'Total ART coverage', 'deaths_estimate']:
		rate_df[f'delta_{col}'] = rate_df[col].diff()
	
	print('\nRates of change (first 10 rows):')
	display(rate_df.head(10))
```

```python
# Diagnostics â Overfitting checks for model forms
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
from sklearn.model_selection import TimeSeriesSplit

# Helper metrics
def compute_metrics(y_true, y_pred, num_predictors):
	"""Return dict with R2, adjR2, AIC, BIC, RSS, DW."""
	y_true = np.asarray(y_true, dtype=float)
	y_pred = np.asarray(y_pred, dtype=float)
	n = y_true.shape[0]
	if n < max(3, num_predictors + 2):
		return {
			'R2': np.nan,
			'adjR2': np.nan,
			'AIC': np.nan,
			'BIC': np.nan,
			'RSS': np.nan,
			'DW': np.nan,
		}
	res = y_true - y_pred
	RSS = float(np.sum(res**2))
	R2 = float(r2_score(y_true, y_pred))
	k = num_predictors + 1  # include intercept
	adjR2 = float(1 - (1 - R2) * (n - 1) / (n - k)) if (n - k) > 0 else np.nan
	# Gaussian likelihood approximation
	sigma2 = RSS / n
	AIC = float(n * np.log(sigma2) + 2 * k) if sigma2 > 0 else np.nan
	BIC = float(n * np.log(sigma2) + k * np.log(n)) if sigma2 > 0 else np.nan
	# DurbinâWatson for autocorrelation in residuals (sequence assumed chronological)
	DW = float(np.sum(np.diff(res)**2) / np.sum(res**2)) if np.sum(res**2) > 0 and len(res) > 1 else np.nan
	return {'R2': R2, 'adjR2': adjR2, 'AIC': AIC, 'BIC': BIC, 'RSS': RSS, 'DW': DW}

# Cross-validation with forward-chaining
def tscv_r2_scores(X, y, model_kind, max_splits=5):
	X = np.asarray(X, dtype=float)
	y = np.asarray(y, dtype=float)
	n = len(y)
	n_splits = int(min(max_splits, max(2, n - 2)))
	if n_splits < 2:
		return np.nan, np.nan
	ts = TimeSeriesSplit(n_splits=n_splits)
	scores = []
	for train_idx, test_idx in ts.split(X):
		X_tr, X_te = X[train_idx], X[test_idx]
		y_tr, y_te = y[train_idx], y[test_idx]
		if model_kind == 'linear':
			m = LinearRegression().fit(X_tr, y_tr)
			yhat = m.predict(X_te)
			scores.append(r2_score(y_te, yhat))
		elif model_kind == 'log':
			mask_tr = X_tr.flatten() > 0
			mask_te = X_te.flatten() > 0
			if mask_tr.sum() < 2 or mask_te.sum() < 1:
				continue
			m = LinearRegression().fit(np.log(X_tr[mask_tr]), y_tr[mask_tr])
			yhat = m.predict(np.log(X_te[mask_te]))
			scores.append(r2_score(y_te[mask_te], yhat))
		elif model_kind == 'exp':
			mask_tr = y_tr > 0
			mask_te = y_te > 0
			if mask_tr.sum() < 2 or mask_te.sum() < 1:
				continue
			m = LinearRegression().fit(X_tr[mask_tr], np.log(y_tr[mask_tr]))
			yhat = np.exp(m.predict(X_te[mask_te]))
			scores.append(r2_score(y_te[mask_te], yhat))
		elif model_kind == 'quad':
			pf = PolynomialFeatures(degree=2, include_bias=False)
			X_trq = pf.fit_transform(X_tr)
			X_teq = pf.transform(X_te)
			m = LinearRegression().fit(X_trq, y_tr)
			yhat = m.predict(X_teq)
			scores.append(r2_score(y_te, yhat))
		else:
			continue
	if len(scores) == 0:
		return np.nan, np.nan
	return float(np.nanmean(scores)), float(np.nanstd(scores))

# Run diagnostics for both outcomes
if analysis_df.empty:
	print('Diagnostics skipped: analysis_df is empty.')
else:
	results = []
	for y_col in ['HIV prevalence (%)', 'deaths_estimate']:
		df = analysis_df.sort_values('Year').reset_index(drop=True)
		X = df[['Total ART coverage']].values
		y = df[y_col].values

		# Linear
		lin = LinearRegression().fit(X, y)
		yhat_lin = lin.predict(X)
		m_lin = compute_metrics(y, yhat_lin, num_predictors=1)
		cv_lin_mean, cv_lin_std = tscv_r2_scores(X, y, 'linear')
		results.append({'Outcome': y_col, 'Model': 'Linear', **m_lin, 'CV_R2_mean': cv_lin_mean, 'CV_R2_sd': cv_lin_std})

		# Logarithmic (X>0)
		mask_pos = (X.flatten() > 0)
		if mask_pos.sum() >= 3:
			X_log = np.log(X[mask_pos])
			y_log = y[mask_pos]
			logm = LinearRegression().fit(X_log, y_log)
			yhat_log = logm.predict(X_log)
			m_log = compute_metrics(y_log, yhat_log, num_predictors=1)
			cv_log_mean, cv_log_std = tscv_r2_scores(X, y, 'log')
			results.append({'Outcome': y_col, 'Model': 'Logarithmic', **m_log, 'CV_R2_mean': cv_log_mean, 'CV_R2_sd': cv_log_std})
		else:
			results.append({'Outcome': y_col, 'Model': 'Logarithmic', 'R2': np.nan, 'adjR2': np.nan, 'AIC': np.nan, 'BIC': np.nan, 'RSS': np.nan, 'DW': np.nan, 'CV_R2_mean': np.nan, 'CV_R2_sd': np.nan})

		# Exponential (y>0)
		mask_ypos = (y > 0)
		if mask_ypos.sum() >= 3:
			expm = LinearRegression().fit(X[mask_ypos], np.log(y[mask_ypos]))
			yhat_exp = np.exp(expm.predict(X[mask_ypos]))
			m_exp = compute_metrics(y[mask_ypos], yhat_exp, num_predictors=1)
			cv_exp_mean, cv_exp_std = tscv_r2_scores(X, y, 'exp')
			results.append({'Outcome': y_col, 'Model': 'Exponential', **m_exp, 'CV_R2_mean': cv_exp_mean, 'CV_R2_sd': cv_exp_std})
		else:
			results.append({'Outcome': y_col, 'Model': 'Exponential', 'R2': np.nan, 'adjR2': np.nan, 'AIC': np.nan, 'BIC': np.nan, 'RSS': np.nan, 'DW': np.nan, 'CV_R2_mean': np.nan, 'CV_R2_sd': np.nan})

		# Quadratic
		pf = PolynomialFeatures(degree=2, include_bias=False)
		X_quad = pf.fit_transform(X)
		quad = LinearRegression().fit(X_quad, y)
		yhat_quad = quad.predict(X_quad)
		m_quad = compute_metrics(y, yhat_quad, num_predictors=2)
		cv_quad_mean, cv_quad_std = tscv_r2_scores(X, y, 'quad')
		results.append({'Outcome': y_col, 'Model': 'Quadratic', **m_quad, 'CV_R2_mean': cv_quad_mean, 'CV_R2_sd': cv_quad_std})

	res_df = pd.DataFrame(results)
	# Order columns for readability
	cols = ['Outcome', 'Model', 'R2', 'adjR2', 'AIC', 'BIC', 'RSS', 'DW', 'CV_R2_mean', 'CV_R2_sd']
	res_df = res_df[cols]
	print('Model diagnostics (lower AIC/BIC better; DWâ2 implies low autocorrelation):')
	display(res_df)
```

### How to detect overfitting here
- **Adjusted RÂ²**: Penalizes extra parameters. If adjusted RÂ² is similar to or below plain RÂ² (or below the simpler modelâs adjusted RÂ²), added curvature likely doesnât generalize.
- **AIC/BIC**: Lower is better after penalizing complexity. If quadratic/log has lower AIC/BIC than linear, that supports the extra curvature.
- **Time-series CV (forward chaining)**: We trained on early years and tested on later years. If cross-validated RÂ² for quadratic/log stays high (and above linear), they are less likely to overfit.
- **Residual diagnostics (DurbinâWatson)**: DWâ2 suggests residuals arenât serially autocorrelated. Strong autocorrelation hints at missed time dynamics.

Read the diagnostics table printed above: prefer the model with lower AIC/BIC, higher adjusted RÂ², and higher CV RÂ² (with small CV sd). If quadratic/log wins all three, overfitting risk is lower; if only in-sample RÂ² is higher but AIC/BIC and CV worsen, thatâs overfitting.

### Step D â Deeper explanation
- **Why lags?** ART coverage at year t might affect outcomes with delay. We aligned ART(t) to outcomes at t+k (k=0,1,2). Rising r with k for prevalence suggests delayed impact; deaths respond quicker as treatment lowers mortality promptly.

- **Prevalence dynamics**: Prevalence = people living with HIV / population. ART reduces deaths and increases life expectancy for PLHIV, which can raise or stabilize prevalence even if new infections fall. Hence strong positive ARTâprevalence correlations and near-flat post-2004 slope.

- **Deaths dynamics**: ART directly reduces mortality among PLHIV. Thatâs why ARTâdeaths correlation is strongly negative and improves with a short lag.

- **Piecewise slopes (pre/post-2004)**: 2004 approximates scale-up. Pre-2004 slopes are confounded by the epidemicâs natural rise while ART was minimal, so positive/odd slopes arenât causal. Post-2004 slopes are more interpretable as associations during the treatment era:
  - HIV prevalence: ~0 slope per +1% ART â stabilization.
  - AIDS deaths: â â3,014 per +1% ART â substantial declines associated with scale-up.

- **Rates of change (Î per year)**: Inspect `delta_*` columns. If ART grows while `delta_deaths` is negative and `delta_prevalence` ~0, that matches the story: ART cuts deaths; prevalence stabilizes.

- **Cautions provided by ChatGPT**: These are bivariate associations. Unmeasured factors (testing, prevention, policy, COVID) and autocorrelation can bias results. For causal claims, consider controls (incidence, condom use, PMTCT), difference-in-differences, or structural models.
